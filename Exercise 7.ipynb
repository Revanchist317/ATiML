## Assignment 7.1

Describe in a few words the goal of Semi-Supervised Learning. What problems are there? What challenges?

Annotating data and creating labels is very costly and time-consuming, often requiring expert knowledge. Therefore, supervised learning might only leverage a small portion of the labelled data, albeit a larger unlabelled pool is available. On the other hand, unsupervised learning dispenses with the labels altogether. This is disadvantageous since we often have _some_ domain knowledge, albeit not perfect. Therefore, the goal of SSL is to leverage the limited domain knowledge to improve the performance of a model, utilising both labelled and unlabelled data points. 

There are a few unique challenges in this domain of ML, notably the underlying assumptions regarding the distribution of the data wherein the models assume unlabelled data close to the labelled data will most likely have the same labels (continuity assumption). Relatedly, early mistakes in classifying unlabelled instances might propagate in some instances. Additionally, there is no closed form solution (often) rendering the problem at hand susceptible to non-convergence.
